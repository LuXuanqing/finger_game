```python
import cv2 as cv
```

```python
cap = cv.VideoCapture(0)
```

```python
def ka():
    while True:
        _, f = cap.read()
        cv.imshow('cap',f)
        print(cap.get(28), cap.get(39))
        if cv.waitKey(1) == ord('q'):
            break
    cv.destroyAllWindows()
```

```python
ka()
```
